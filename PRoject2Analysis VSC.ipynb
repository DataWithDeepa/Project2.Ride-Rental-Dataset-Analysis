# Step 7: ** Data Wrangling **: 
#### Merging datasets using a common key. Dropping unnecessary columns to keep only relevant data.

```python
print(df1.columns)  
```

```python
print(df2.columns)
```

```python
print(df3.columns)
```

```python
df1.rename(columns={'Id': 'id', 'ID': 'id', 'id_number': 'id'}, inplace=True)
```

```python
df2.rename(columns={'Id': 'id', 'ID': 'id', 'id_number': 'id'}, inplace=True)
```

```python
df3.rename(columns={'Id': 'id', 'ID': 'id', 'id_number': 'id'}, inplace=True)
```

```python
if 'id' not in df1.columns:
    df1['id'] = range(1, len(df1) + 1)
```

```python
if 'id' not in df2.columns:
    df2['id'] = range(1, len(df2) + 1)
```

```python
if 'id' not in df3.columns:
    df3['id'] = range(1, len(df3) + 1)
```

```python
merged_df = df1.merge(df2, on='id', how='outer').merge(df3, on='id', how='outer')
```

```python
print(merged_df.head())
print(merged_df.isnull().sum())  # Missing values check karein
```

# Step 8 ** Final Steps: Step  :  to Complete Project **

#### Handle Missing Values

#### Fill missing values for numerical columns with median

```python
merged_df.fillna(merged_df.median(numeric_only=True), inplace=True)
```

#### Fill missing values for categorical columns with mode

```python
pd.set_option('future.no_silent_downcasting', True)
```

```python
for col in merged_df.select_dtypes(include=['object']).columns:
    merged_df[col] = merged_df[col].fillna(merged_df[col].mode().iloc[0])
```

#### Handling missing values (Forward Fill and Drop Columns if too many missing)

```python
threshold = 0.5 * len(df)  # 50% missing threshold
```

```python
df.dropna(axis=1, thresh=threshold, inplace=True)  # Drop columns with >50% missing values
```

```python
df = df.infer_objects(copy=False)  # To correct Object dtype 
```

```python
df.bfill(inplace=True)
```

# { Create final visualization }

```python
# To Make Figure & Axes 
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Overall Title for all subplots
plt.suptitle("Final Visual Representation of Ride Rental Analysis", fontsize=16, fontweight='bold')

# Temperature vs Rentals
sns.scatterplot(x=df['temp'], y=df['cnt'], alpha=0.6, ax=axes[0, 0])
axes[0, 0].set_title("Temperature vs Ride Rentals")

# Weather Impact
sns.boxplot(x='weathersit', y='cnt', data=df, hue='weathersit', palette='coolwarm', ax=axes[0, 1])
axes[0, 1].set_title("Impact of Weather on Ride Rentals")

# Rentals per Hour
sns.lineplot(x='hr', y='cnt', data=df, marker='o', color='red', ax=axes[1, 0])
axes[1, 0].set_title("Ride Rentals by Hour of the Day")

# Rentals by Day of the Week
sns.barplot(x='weekday', y='cnt', hue='weekday', data=df, palette='viridis', ax=axes[1, 1])
axes[1, 1].set_title("Total Rentals by Day of the Week")

# Adjust layout properly
plt.tight_layout(rect=[0, 0, 1, 0.96]) 

# Show final visualization
plt.show()
```
